```python
import sys

# Имя файла для хранения задач
TODO_FILE = "todo.txt"

def display_tasks():
    try:
        with open(TODO_FILE, "r") as f:
            tasks = f.readlines()
            if not tasks:
                print("Список пуст.")
            for i, task in enumerate(tasks, 1):
                print(f"{i}. {task.strip()}")
    except FileNotFoundError:
        print("Список пуст (файл еще не создан).")

def add_task(task):
    with open(TODO_FILE, "a") as f:
        f.write(task + "\n")
    print(f"Добавлено: {task}")

def remove_task(index):
    try:
        with open(TODO_FILE, "r") as f:
            tasks = f.readlines()

        task_to_remove = tasks.pop(index - 1)

        with open(TODO_FILE, "w") as f:
            f.writelines(tasks)
        print(f"Удалено: {task_to_remove.strip()}")
    except (IndexError, ValueError):
        print("Ошибка: Неверный номер задачи.")

if __name__ == "__main__":
    args = sys.argv[1:]

    if not args or args[0] == "list":
        display_tasks()
    elif args[0] == "add" and len(args) > 1:
        add_task(" ".join(args[1:]))
    elif args[0] == "remove" and len(args) > 1:
        remove_task(int(args[1]))
    else:
        print("Использование:\n  python todo.py list\n  python todo.py add [задача]\n  python todo.py remove [номер]")
```
